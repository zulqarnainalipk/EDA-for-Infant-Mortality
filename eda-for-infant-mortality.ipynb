## Dataset Information

This dataset is a pandas DataFrame with 48 entries and 9 columns. It provides information on infant mortality rates and related statistics broken down by year and maternal race or ethnicity.

### Data Columns

1. **Year**: Year of the data entry.
2. **Maternal Race or Ethnicity**: Race or ethnicity of the mother.
3. **Infant Mortality Rate**: The number of infant deaths per 1,000 live births.
4. **Neonatal Mortality Rate**: The number of neonatal deaths (within the first 28 days of life) per 1,000 live births.
5. **Postneonatal Mortality Rate**: The number of postneonatal deaths (from 28 days to 1 year of age) per 1,000 live births.
6. **Infant Deaths**: Total number of infant deaths.
7. **Neonatal Infant Deaths**: Total number of neonatal infant deaths.
8. **Postneonatal Infant Deaths**: Total number of postneonatal infant deaths.
9. **Number of Live Births**: Total number of live births.

### Data Types

- **int64**: Integer data type.
- **float64**: Floating-point numeric data type.
- **object**: Categorical data type (specifically, text strings indicating race or ethnicity).

### Missing Values

•	Year                                                 0
•	Materal Race or Ethnicity                0
•	Infant Mortality Rate                       8
•	Neonatal Mortality Rate                  8
•	Postneonatal Mortality Rate            9
•	Infant Deaths                                   8
•	Neonatal Infant Deaths                    8
•	Postneonatal Infant Deaths              8
•	Number of Live Births                      0

### Filling of Missing Values
We fill in missing values by using the average values of each column. This helps keep our data consistent and prevents any big changes or errors. It's a good way to fill in missing information without causing problems or messing up the data.


```python
import warnings
warnings.filterwarnings('ignore')  # Ignore warnings

import numpy as np  # Numerical operations
import pandas as pd  # Data manipulation
import matplotlib.pyplot as plt  # Data visualization
```

```python
df=pd.read_csv('/kaggle/input/infant-mortality/infant-mortality-1.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python


# Assuming your DataFrame is named 'data'
columns_with_missing_values = ['Infant Mortality Rate', 'Neonatal Mortality Rate', 
                                'Postneonatal Mortality Rate', 'Infant Deaths', 
                                'Neonatal Infant Deaths', 'Postneonatal Infant Deaths']

# Fill null values with mean for specified columns
df[columns_with_missing_values] = df[columns_with_missing_values].fillna(df[columns_with_missing_values].mean())

# Verify if null values are filled
print(df.isna().sum())
```

```python
df.info()
```

# Bar Plots for Mortality Rates by Ethnicity
---

```python
# Bar Plots for Mortality Rates by Ethnicity
mortality_rates = ['Infant Mortality Rate', 'Neonatal Mortality Rate', 'Postneonatal Mortality Rate']

plt.figure(figsize=(10, 6))

plt.bar(df['Materal Race or Ethnicity'], df[mortality_rates[0]], alpha=0.7, label=mortality_rates[0],color='blue')
plt.bar(df['Materal Race or Ethnicity'], df[mortality_rates[1]], alpha=0.7, label=mortality_rates[1],color='green', bottom=df[mortality_rates[0]])
plt.bar(df['Materal Race or Ethnicity'], df[mortality_rates[2]], alpha=0.7, label=mortality_rates[2],color='orange', bottom=df[mortality_rates[0]] + df[mortality_rates[1]])

plt.xlabel('Ethnicity')
plt.ylabel('Mortality Rate')
plt.title('Mortality Rates by Ethnicity')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Pie Chart for Ethnicity Distribution
---

```python
# Pie Chart for Ethnicity Distribution
ethnicity_counts = df['Materal Race or Ethnicity'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(ethnicity_counts, labels=ethnicity_counts.index,startangle=140, autopct='%1.1f%%',colors=['gold', 'green','blue','lightcoral', 'lightskyblue', 'lightpink'])
plt.title('Ethnicity Distribution')
plt.show()
```

# Line Plot for Infant Mortality Rate Over Time
---

```python

# Pivot the DataFrame
df2 = df.pivot(index='Year', columns='Materal Race or Ethnicity', values='Infant Mortality Rate')

# Plot the pivoted DataFrame
plt.figure(figsize=(10, 10))
df2.plot(kind='line', marker='o')

plt.xlabel('Year')
plt.ylabel('Infant Mortality Rate')
plt.title('Infant Mortality Rate Over Time by Ethnicity')
plt.legend(title='Ethnicity')
plt.tight_layout()
plt.show()
```

# Bar plot of Average Yearly Deaths by Ethnicity
---

```python
df['Total Deaths'] = df['Infant Deaths'] + df['Neonatal Infant Deaths'] + df['Postneonatal Infant Deaths']

# Calculate average number of deaths per year for each ethnicity
average_deaths = df.groupby('Materal Race or Ethnicity')['Total Deaths'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
average_deaths.plot(kind='bar', color='coral')
plt.xlabel('Materal Race or Ethnicity')
plt.ylabel('Average Number of Deaths')
plt.title('Average Yearly Deaths by Ethnicity')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Print the average number of deaths (optional)
print(average_deaths)
```

# Bar plot of Average Yearly Live Births by Ethnicity
---

```python
# Calculate average number of live births per year for each ethnicity
average_live_births = df.groupby('Materal Race or Ethnicity')['Number of Live Births'].mean()


# Create a bar chart to visualize the data
plt.figure(figsize=(10, 6))
average_live_births.plot(kind='bar', color='skyblue')
plt.xlabel('Materal Race or Ethnicity')
plt.ylabel('Average Number of Live Births')
plt.title('Average Yearly Live Births by Ethnicity')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Scatter Plot for Infant Mortality Rate vs. Number of Live Births
---

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Number of Live Births'], df['Infant Mortality Rate'], alpha=0.5)
plt.xlabel('Number of Live Births')
plt.ylabel('Infant Mortality Rate')
plt.title('Infant Mortality Rate vs. Number of Live Births')
plt.show()
```
